# Відповіді на контрольні запитання

1. Інтерпретатор і компілятор - це два різних підходи до виконання програмного коду. Python використовує інтерпретатор.

Основна відмінність між ними полягає в тому, як вони обробляють програмний код:

### Інтерпретатор:

Інтерпретатор виконує програмний код по одній інструкції на раз.
Він читає вихідний код програми і виконує його без попереднього перетворення у машинний код.
Код виконується поетапно, рядок за рядком.
Якщо в процесі виконання виникають помилки, вони виявляються на льоту.
Python - це приклад інтерпретованої мови програмування.

### Компілятор:

Компілятор перетворює вихідний код програми у машинний код або код на низькорівневій мові програмування перед виконанням.
Після компіляції програми виконання відбувається швидше, оскільки машинний код може бути виконаний безпосередньо.
Якщо в процесі компіляції виявляються помилки, то програму не можна буде виконувати, поки помилки не будуть виправлені.
Прикладами мов програмування, які використовують компіляцію, є C, C++, та Java (яка використовує гібридний підхід з компіляцією та інтерпретацією).
Отже, Python - це інтерпретована мова програмування, оскільки код виконується по одній інструкції за раз і не потребує попередньої компіляції у машинний код.

2. PyCharm - це інтегроване середовище розробки (IDE) для мови програмування Python, яке розробляється компанією JetBrains. Це одна з найпопулярніших IDE для розробки на Python, і вона має численні функції та інструменти для полегшення роботи розробників Python.


3. Система контролю версій (СКВ) - це програмне забезпечення, яке призначене для відстеження змін у вихідному коді та історії робочих файлів. Головна мета СКВ полягає в збереженні історії змін, спрощенні спільної роботи декількох розробників над одним проектом, відновленні попередніх версій коду та управлінні конфліктами внесених змін. Принципова відмінність між СКВ SVN (Apache Subversion) та GIT:


Розподілена vs. Централізована система:

GIT є розподіленою системою контролю версій, що означає, що кожен розробник має локальний репозиторій і може працювати незалежно від мережевого підключення.
SVN є централізованою системою, де центральний сервер містить всю історію проекту і розробники взаємодіють з цим сервером.
Модель гілок (Branching) та об'єднання (Merging):

GIT має потужну модель гілок та об'єднання, що дозволяє розробникам створювати галузі для роботи над функціоналом та об'єднувати їх без зусиль.
SVN підтримує гілки, але об'єднання може бути менш зручним та призводити до конфліктів.
Ідентифікація змін:

GIT і SVN використовують імена авторів для ідентифікації змін, але GIT дозволяє більше гнучкості в цьому плані.
Ефективність та швидкодія:

GIT зазвичай працює швидше завдяки розподіленій моделі та компактному збереженню даних.
SVN може бути повільнішим, особливо при великих проектах.
Обираючи між SVN і GIT, вибір залежить від конкретних потреб та типу проекту. GIT дедалі більше стає стандартом в сучасній розробці завдяки своїм перевагам у роботі з гілками та розподіленим робочим процесом.

4. Команда `git add .` у системі контролю версій Git використовується для додавання змінених або нових файлів до стейджу (проміжного області). Після виконання цієї команди зміни, які були внесені до робочого каталогу, додаються до стейджу, і вони готові для включення до наступного коміту (зафіксування змін в історії версій).

5. Команда git commit у системі контролю версій Git використовується для фіксації (зафіксування) змін у репозиторії та створення нового коміту (версії) з цими змінами. Команда git commit - це ключовий етап в роботі з Git, оскільки вона додає зміни до історії версій та робить їх доступними для спільної роботи та відстеження.

6. Для розміщення локальної копії Git-репозиторію на віддаленому сервері, як, наприклад, GitHub, використовується команда git push. Ця команда відправляє локальні зміни до віддаленого репозиторію на сервері.
